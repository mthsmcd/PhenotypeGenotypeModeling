```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
num_exp = 20
data = np.array([np.loadtxt(f'outputs/meanf_{i}.dat') for i in range(num_exp)])
num_gens = data.shape[1]
x_plot = range(num_gens)

plt.figure()
[plt.plot(x_plot, data[i]) for i in range(num_exp)];
plt.xlabel('Generation');
plt.ylabel('Mean Population Fitness');
plt.title(f'All {num_exp} experiments');

mean_fit = np.mean(data, axis=0)
std_dev = np.std(data, axis=0)
plt.figure()
plt.plot(x_plot, mean_fit, 'b')
plt.plot(x_plot, mean_fit + std_dev, 'r')
plt.plot(x_plot, mean_fit - std_dev, 'r')
plt.xlabel('Generation');
plt.ylabel('Mean Population Fitness');
plt.title(f'Mean fitness of {num_exp} experiments');
plt.ylim((0,1.0))
plt.grid(True, 'both', axis='y', linestyle='--', color='k')

#annotation config. to display maximum fit information
xmax = x_plot[np.argmax(mean_fit)]
ymax = mean_fit.max()
text= r"$\bf{Maximum}$ $\bf{\mu}$"+"\n"+r"$\mu$ = {:.2f} at gen. = {:.0f}".format(ymax, xmax)
bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=70")
kw = dict(xycoords='data', textcoords="axes fraction", horizontalalignment="center",
        arrowprops=arrowprops, bbox=bbox_props, va="top")
plt.annotate(text, xy=(xmax, ymax), xytext=(0.7,0.55), **kw)
plt.legend((r'Mean: $\mu$',r'Mean $\pm$ Std. Dev: $\mu \pm \sigma$'))
```
